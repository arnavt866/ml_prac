Data Preprocessing

```python
# Load the two CSVs
df_true = pd.read_csv("/content/true.csv")
df_fake = pd.read_csv("/content/fake.csv")

# Add a label column so we know which row is True/Fake
df_true["label"] = "TRUE"
df_fake["label"] = "FAKE"

# Merge them into one DataFrame
df = pd.concat([df_true, df_fake], ignore_index=True)

# Save as one combined CSV (optional)
df.to_csv("/content/news_dataset.csv", index=False)

print("Merged dataset shape:", df.shape)
```
